```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use("seaborn-v0_8-whitegrid")
```

```python
print(plt.style.available)
```

```python
iris = pd.read_csv("iris.csv",sep=",")
iris.head(5)
```

```python
fig = plt.figure()
ax = plt.axes()

# Doğru şekilde sütunlara erişerek scatter plot çiz
plt.plot(iris.sepal_length, iris.sepal_width, marker="o", linestyle="None")

# Eksen başlıkları ve grafik başlığı
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("İris Çiçekleri: Sepal Ölçüleri")
plt.show()
```

```python
plt.scatter(iris.sepal_length,iris.petal_length,s=55,c="m",marker="v",edgecolor="black",linewidths=1,alpha=1)
```

```python
renk = np.random.randint(0,10,150)
boyut = np.random.randint(50,200,150)
plt.scatter(iris.sepal_length,iris.petal_length,s=boyut,c=renk,cmap="Reds",alpha=1)
cbar=plt.colorbar()
cbar.set_label("Seviyeler")
plt.xlabel("sepal boy")
plt.ylabel("petal boy")
plt.title("iris çiçeği sepaal boy petal boy")
```
